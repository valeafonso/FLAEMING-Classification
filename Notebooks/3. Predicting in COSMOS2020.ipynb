```python
dfscore = []
dfproba= []
for k in range(len(testing_in_wild)):
    
    pred_sum = preds1_lgbm[k] + preds2_lgbm[k] + preds3_lgbm[k] + preds4_lgbm[k] + preds5_lgbm[k] + preds1_xg[k] + preds2_xg[k] + preds3_xg[k] + preds4_xg[k] + preds5_xg[k] +preds1_cb[k] + preds2_cb[k] + preds3_cb[k] + preds4_cb[k] + preds5_cb[k]

    dfscore.append(pred_sum)
    prob = (preds_proba1_lgbm[k]+preds_proba2_lgbm[k]+preds_proba3_lgbm[k]+preds_proba4_lgbm[k]+preds_proba5_lgbm[k]+
        preds_proba1_xg[k]+preds_proba2_xg[k]+preds_proba3_xg[k]+preds_proba4_xg[k]+preds_proba5_xg[k]+
        preds_proba1_cb[k]+preds_proba2_cb[k]+preds_proba3_cb[k]+preds_proba4_cb[k]+preds_proba5_cb[k])/15
    dfproba.append(prob)

testing_in_wild['Times_Predicted (/15)'] = dfscore
testing_in_wild['Average_Predict_Prob'] = dfproba
```
